```python
from sklearn.metrics.pairwise import euclidean_distances, paired_distances, rbf_kernel
import scipy as scp
class multivariate_normal():
    '''
    Input: 
        mu, sigma
    Return:
    '''
    def __init__(self,mu,sigma=None):
        #super(multivariate_normal).__init__()
        
        #[x1,x2,x3] -> [x1,x2,x3]
        #[[x1,x2,x3]] -> [x1,x2,x3]
        #[[x1],[x2],[x3]] -> [x1,x2,x3]
        if not isinstance(mu,np.ndarray):
            raise TypeError(f"mu shuld be numpy.ndarray, but got {type(mu)}")
        
        if len(mu.shape)!=1:
            if max(mu.shape) != mu.size:
                raise ValueError(f"shape of mu argument given to the gauss_kernel should be"\
                                 +f"(:,1) or (1,:) or (:,), but got {mu.shape}")#[[x1,x2],[x3,x4]] -> Error
        self.mu = mu.ravel()
        
        if sigma is None:
            self.sigma = np.eye(self.mu.size)
        elif not isinstance(sigma,np.ndarray):
            self.sigma = np.eye(self.mu.size)*sigma
        else:
            if (len(sigma.shape) == 1)&(sigma.size == self.mu.size):
                self.sigma = np.diag(sigma) #[x1,x2,x3] -> diag([x1,x2,x3])
            elif sigma.size != self.mu.size:
                raise ValueError(f"size of sigma argument givent to the gauss_kernel should be as same as the size of mu:{self.mu.size}, but got {sigma.size}")
        
    def pdf(self, x):
        '''
        Input: numpy array x
        Output: pdf
        '''
        if x.size != self.mu.size:
            raise ValueError(f"shape of x argument given to pdf(x) should be as same as the size of mu {self.mu.shape}, but got {x.size}")
        p = scp.stats.multivariate_normal.pdf(x,self.mu,self.sigma)
        return p

class gauss_kernel():
    def __init__(self,sigma):
        self.sigma = sigma
        
    def __call__(self,x,y):
        return self.compute(x,y)
    
    def compute(self, x,y):
        if x.shape != y.shape:
            raise ValueError(f"operands could not be broadcast together with shapes {x.shape} {y.shape}")
        
        p = scp.stats.multivariate_normal.pdf(x,y,self.sigma)
        return p

    def set_sigma(self, sigma):
        self.sigma = sigma

def get_median_distance(d):
    if d.shape[0] == 1:
        raise ValueError(f"samplesize of operand should be >1, but got {d.shape}.")
    K = euclidean_distances(d, squared=True) #データ間のユークリッド距離を総当たり的に求めてる
    sigma = np.median(np.sqrt(K)) #最頻する距離で正規化するためにmedian取ってる
    return sigma

def get_band_width(d, method='median'):
    '''
    Determine bandwidth of gaussian kernel. 
    Method:
        - Median heuristic
        - Silverman heuristic
        - Scott heuristic
        - LSCV
    '''
    
    if method == 'median':
        K = euclidean_distances(d, squared=True) #データ間のユークリッド距離を総当たり的に求めてる
        sigma = np.median(np.sqrt(K)) #最頻する距離で正規化するためにmedian取ってる
        
    elif method in ['silverman', 'scott']:
        factor = 1.06 if method == 'silverman' else 0.9
        distances_from_mean = euclidean_distances(d,[d.mean(axis=0)], squared=False)
        std = np.sqrt(np.square(distances_from_mean).mean())
        sig = min(std,\
                  (np.percentile(distances_from_mean,75)-np.percentile(distances_from_mean,25))/1.34)
        sigma = factor*sig/(d.shape[0]**(1./5))
        
    elif method == 'LSCV':
        raise KeyError('Sorry, LSCV method have not implemented yet.')
    
    else:
        raise ValueError(f"method should be ['median', 'silverman', 'scott', 'LSCV']")
    
    return sigma
        
class KernelMean():
    def __init__(self, x, sigma='median', weights=None):
        self.x = x
        
        self.sigma = sigma
        if isinstance(sigma,str):
            self.sigma = get_band_width(x,sigma)
        
        self.p = np.full(len(x),1./len(x))
        if weights is not None:
            self.p = weights/np.sum(weights)
                
        self.mu_p = lambda x: self._compute_kernel_from_samples(x)
      
    def _compute_kernel_from_samples(self, x):
        kernel = gauss_kernel(self.sigma)
        val = 0
        for sample, p in zip(self.x,self.p):
            val += p*kernel(x,sample)
        return val
        
class KernelHerding():
    def __init__(self, obj_KernelMean):
        if not isinstance(obj_KernelMean, KernelMean):
             raise TypeError(f"init object shuld be KernelMean class, but got {type(obj_KernelMean)}")
        self.KM = obj_KernelMean
        self.samples = []
        
    def supersample(self, sample_size, x_initial_guess=None, options=None, verbose={'clear_display':False, 'display':'notebook'}):
        prob = progbar(sample_size, **verbose)
        h = self.KM.mu_p
        x1 = self._argmax(h, x_initial_guess, options)
        
        samples = [x1]
        for idx in range(sample_size-1):
            h = self._herding_update(h, samples)
            x = self._argmax(h, x_initial_guess, options)
            samples.append(x)
            prob.update(idx+1)
            
        self.herding_h = h
        self.samples +=samples
        del prob
        return samples
    
    def supplement_sample(self, sample_size, x_initial_guess=None, options=None, verbose={'clear_display':False, 'display':'notebook'}):
        prob = progbar(sample_size, **verbose)
        h = self.herding_h
        samples = []
        for idx in range(sample_size-1):
            h = self._herding_update(h, self.samples)
            x = self._argmax(h, x_initial_guess, options)
            samples.append(x)
            prob.update(idx+1)
            
        self.herding_h = h
        self.samples +=samples
        del prob
        return samples
    
    def _argmax(self, h, x_initial_guess=None, options=None):
        minus_h = lambda x: -1.*h(x)
        if options is None:
            options = {'optimizer':'scipy_optim','method':'CG', 'options':{'norm':1e-4, 'gtol':1e-3, 'maxiter':1e+3}}
        if options['optimizer']=='scipy_optim':
            if x_initial_guess is None:
                x_initial_guess = 2*(np.random.rand(2)-0.5)
            optimize_result = scp.optimize.minimize(minus_h, x_initial_guess, method=options['method'], options=options['options'])
            return optimize_result.x
        
        
    def _herding_update(self, h, samples):
        N = float(len(samples))
       #f = lambda x: (1./N)*self.mu_p(x) + ((N-1.)/N)*h(x) - (1./N)*self.kernel(x,samples[-1])
        f = lambda x: (self.mu_p(x) + (N-1.)*h(x) - self.kernel(x,samples[-1]))/N
        return f
    
#     def _objective(trial,)
```
